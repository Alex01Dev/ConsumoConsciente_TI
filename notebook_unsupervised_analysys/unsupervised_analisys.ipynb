### 1. Propuesta de la Aplicación

Este análisis busca encontrar patrones ocultos en productos sustentables para agruparlos según sus características como huella de carbono, precio, reciclabilidad, etc.
La idea es mejorar el sistema de recomendaciones al usuario.

### 2. Elección del mecanismo

Utilizaremos:

K-Means Clustering para agrupar productos.

PCA (Análisis de Componentes Principales) para reducir dimensionalidad y visualizar mejor.

Heatmaps y Scatterplots para interpretar resultados.

### 3. Procesamiento

```python
# Seleccionar variables relevantes
features = ["carbon_footprint", "price", "recyclable_packaging", "local_origin", "sustentabilidad_score"]

X = cc_dataset[features]

# Estandarizar datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled[:5]
```

```python
# Definir número de clusters
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Agregar al dataset
cc_dataset["Cluster"] = clusters

cc_dataset.head()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=clusters, palette="Set2")
plt.title("Clusters de productos (PCA)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(cc_dataset.groupby("Cluster")[features].mean(), annot=True, cmap="YlGnBu")
plt.title("Promedio de características por cluster")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Cluster", y="price", data=cc_dataset, palette="Set2")
plt.title("Distribución de precios por cluster")
plt.show()
```

###  6. Resultados Obtenidos

Los clusters muestran diferencias claras:

Un cluster con productos baratos y sustentables.

Otro con productos de precio medio y buena huella de carbono.

Otro con productos premium pero menos sostenibles.
